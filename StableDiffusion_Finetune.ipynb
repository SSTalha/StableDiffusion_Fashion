```python
from transformers import Trainer, TrainingArguments
from peft import LoraConfig, get_peft_model
from datasets import load_dataset
from transformers import  AutoTokenizer
from diffusers import StableDiffusionPipeline, UNet2DConditionModel, AutoencoderKL, DDPMScheduler
import torch
from torch.utils.data import DataLoader
```

```python
dataset = load_dataset("SaffalPoosh/deepFashion-with-masks")

# Filter and shuffle the dataset to use only 100 images
dataset = dataset['train'].shuffle(seed=42).select(range(100))
dataset = dataset.remove_columns([col for col in dataset.column_names if col not in ["images", "gender", "caption"]])
```

```python
# Preprocess the dataset
def preprocess_function(examples):
    return {
        "image": [image for image in examples["images"]],
        "caption": [caption for caption in examples["caption"]],
        "gender": [gender for gender in examples["gender"]]
    }
```

```python
dataset = dataset.map(preprocess_function, batched=True)
```

```python
model_name = "runwayml/stable-diffusion-v1-5"
model = StableDiffusionPipeline.from_pretrained(model_name, torch_dtype=torch.float16).to("cuda")
```

```python
vae = AutoencoderKL.from_pretrained(model_name, subfolder="vae", torch_dtype=torch.float16).to("cuda")
unet = UNet2DConditionModel.from_pretrained(model_name, subfolder="unet", torch_dtype=torch.float16).to("cuda")
# text_encoder = CLIPTextModel.from_pretrained(model_name, subfolder="text_encoder", torch_dtype=torch.float16).to("cuda")
```

```python
config = LoraConfig(
    r=8,
    lora_alpha=32,
    target_modules=["conv1", "conv2", "conv3","k_proj", "v_proj", "q_proj", "out_proj"],
    lora_dropout=0.1,
    bias="none",
    task_type="text-to-image"
)

unet = get_peft_model(unet, config)
```

```python
scheduler = DDPMScheduler.from_pretrained(model_name, subfolder="scheduler")
```

```python
def collate_fn(batch):
    return {
        "input_ids": torch.stack([tokenizer(caption, return_tensors="pt", truncation=True, padding='max_length', max_length=77).input_ids.squeeze() for caption in batch["caption"]]),
        "images": torch.stack([torch.tensor(image) for image in batch["image"]])
    }

dataloader = DataLoader(dataset, batch_size=4, collate_fn=collate_fn)
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    num_train_epochs=1,
    save_strategy="epoch",
    logging_strategy="epoch"
)
```

```python
trainer = Trainer(
    model=unet,  # Use the modified UNet model
    args=training_args,
    train_dataset=dataset,
)

# Train the model
trainer.train()
```

```python
from transformers import pipeline, AutoTokenizer
from diffusers import StableDiffusionPipeline
import torch
from datasets import load_dataset
import peft
from peft import LoraConfig, get_peft_model, PeftModel
import bitsandbytes as bnb
from transformers import Trainer, TrainingArguments
```

#Prepare Dataset

```python
dataset = load_dataset('SaffalPoosh/deepFashion-with-masks')
dataset = dataset['train'].remove_columns(['pose', 'cloth_type', 'pid', 'mask', 'mask_overlay'])
dataset = dataset.rename_column('caption', 'text')
```

```python
dataset = dataset.shuffle(seed=42).select(range(200))
```

#Import the model

```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
```

#LoRa Config

```python
lora_config = LoraConfig(
    r=4,
    lora_alpha=16,
    target_modules=["conv1", "conv2", "conv3","k_proj", "v_proj", "q_proj", "out_proj", "fc1", "fc2"],
    lora_dropout=0.1
)
```

```python
pipe.text_encoder = get_peft_model(pipe.text_encoder, lora_config)
pipe.text_encoder = pipe.text_encoder.half()
def convert_to_8bit(model):
    for name, module in model.named_children():
        if isinstance(module, torch.nn.Linear):
            input_features = module.in_features
            output_features = module.out_features
            new_module = bnb.nn.Linear8bitLt(input_features, output_features)
            model._modules[name] = new_module
        else:
            convert_to_8bit(module)

convert_to_8bit(pipe.text_encoder)

# Print the modified model to verify
print(pipe.text_encoder)
```

```python
if torch.cuda.is_available():
    pipe.text_encoder = pipe.text_encoder.cuda()
    unet = pipe.unet.cuda()

# Print the trainable parameters to verify
for name, param in pipe.text_encoder.named_parameters():
    if param.requires_grad:
        print(f"Text Encoder Trainable parameter: {name}")

print("LoRa model for Text Encoder successfully configured and trainable parameters set.")
```

```python
# Define Custom Dataset
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, dataset):
        self.dataset = dataset

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]
        image = item['images']
        text = item['text']
        return {
            'pixel_values': pipe.vae.encode(image.unsqueeze(0)).latent_dist.sample().squeeze(0),
            'input_ids': pipe.tokenizer(text, return_tensors="pt").input_ids.squeeze(0)
        }

train_dataset = CustomDataset(dataset)
```

```python
# Training Arguments
training_args = TrainingArguments(
    output_dir="./stable-diffusion-finetuned",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    fp16=True,
    save_steps=10_000,
    save_total_limit=2,
    gradient_accumulation_steps=4,
    logging_steps=100,
    remove_unused_columns=False,
    optim="paged_adamw_8bit",
    push_to_hub=False
)
```

```python
class TextToImageTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False):
        pixel_values = inputs['pixel_values']
        input_ids = inputs['input_ids']
        outputs = model(input_ids=input_ids, pixel_values=pixel_values)
        loss = outputs.loss
        return (loss, outputs) if return_outputs else loss

# Initialize and Start Training
trainer = TextToImageTrainer(
    model=pipe.text_encoder,
    args=training_args,
    train_dataset=train_dataset
)

trainer.train()
```
